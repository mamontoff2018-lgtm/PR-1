## Задания


```python
import math

X = [10, 20, 20, 30, 45, 50]
Y = [12, 22, 24, 35, 38, 55]

# ЗАДАНИЕ 1
class DataPro:
    def __init__(self, data):
        self.data = data
        self.n = len(data)
        self.mean_val = sum(data) / self.n
        self.median_val = self._median()
        self.stdev_val = self._stdev()
        self.cv_val = (self.stdev_val / self.mean_val * 100) if self.mean_val != 0 else 0
        self.z_vals = [(x - self.mean_val) / self.stdev_val if self.stdev_val != 0 else 0 for x in data]

    def _median(self):
        s = sorted(self.data)
        mid = self.n // 2
        if self.n % 2 == 0:
            return (s[mid - 1] + s[mid]) / 2
        return s[mid]

    def _stdev(self):
        m = self.mean_val
        var = sum((x - m) ** 2 for x in self.data) / self.n
        return math.sqrt(var)

    def report(self):
        print("Отчёт DataPro")
        print(f"Данные: {self.data}")
        print(f"Среднее: {self.mean_val:.4f}")
        print(f"Медиана: {self.median_val:.4f}")
        print(f"Коэффициент вариации: {self.cv_val:.2f}%")
        print("Z-оценки:")
        for i, (val, z) in enumerate(zip(self.data, self.z_vals)):
            print(f"  x[{i}] = {val} -> z = {z:.4f}")

# ЗАДАНИЕ 2
def get_correlation(x, y):
    mean_x = sum(x) / len(x)
    mean_y = sum(y) / len(y)

    numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y))
    den_x = sum((xi - mean_x)**2 for xi in x)
    den_y = sum((yi - mean_y)**2 for yi in y)

    return numerator / math.sqrt(den_x * den_y)

# ЗАДАНИЕ 3
def simple_hist(data, bins=3):
    low, high = min(data), max(data)
    step = (high - low) / bins
    intervals = [0] * bins

    for val in data:
        idx = int((val - low) / step)
        if idx == bins:
            idx = bins - 1
        intervals[idx] += 1

    total = len(data)
    print(f"Гистограмма (bins={bins}):")
    for i, count in enumerate(intervals):
        left = low + i * step
        right = low + (i + 1) * step
        pct = (count / total) * 100
        print(f"Интервал {i+1} [{left:.2f}, {right:.2f}]: {pct:.1f}%")

# ДЕМОНСТРАЦИЯ
if __name__ == "__main__":
    # Задание 1
    dp = DataPro(X)
    dp.report()

    # Задание 2
    r_before = get_correlation(X, Y)
    print(f"Корреляция до выброса: {r_before:.4f}")

    X_out = X + [1000]
    Y_out = Y + [0]
    r_after = get_correlation(X_out, Y_out)
    print(f"Корреляция после выброса (X=1000, Y=0): {r_after:.4f}")
    print("Вывод: одна экстремальная точка сильно искажает коэффициент Пирсона.\n")

    # Задание 3
    print("Исходная гистограмма X с 3 интервалами:")
    simple_hist(X)
    print("\nГистограмма X с 5 интервалами:")
    simple_hist(X, bins=5)
```

# Контрольные вопросы

1. Почему при расчёте корреляции мы делим на корень из произведения сумм квадратов отклонений?

    Деление на √(∑(x_i − x̄)² · ∑(y_i − ȳ)²) нормирует ковариацию, приводя её в диапазон [‑1, 1]. Благодаря этому коэффициент Пирсона безразмерен и не зависит от масштаба измеряемых величин (например, можно коррелировать рост в сантиметрах с весом в килограммах).

2. В какой ситуации мода будет полезнее среднего?

    Мода предпочтительна для категориальных данных (цвет, марка), при мультимодальных распределениях, а также при наличии сильных выбросов — в отличие от среднего, мода не смещается экстремальными значениями и показывает самое типичное значение.

3. Корреляция r = –0.95. О чём это говорит?

    О сильной отрицательной линейной связи: когда одна переменная растёт, другая почти линейно убывает. Точки диаграммы рассеяния плотно группируются вокруг прямой с отрицательным наклоном.

4. Как с помощью Z-оценки определить выброс?

    Z‑оценка показывает, на сколько стандартных отклонений значение удалено от среднего. По правилу трёх сигм значение считают выбросом, если |z| > 3. В менее строгих случаях используют порог |z| > 2 (приблизительно 5 % самых экстремальных наблюдений в нормальном распределении).
